## 3. Create a bar chart of average salary by title.

```python
#Join salaries and titles using employees dataframe as intermediate.
#merge employees and salaries, inner join, 
empl_slr = employees.merge(salaries, on='emp_no')
empl_slr
```

```python
#merge title dataframe to the brand new dataframe created
est = empl_slr.merge(titles, left_on='emp_title_id', right_on='title_id', how='outer')

est
```

```python
#group by title and then get the mean of salaries by title

employees_grouped_by_title = est.groupby(['title'])['salary'].mean()
employees_grouped_by_title
```

```python
# plot

employees_grouped_by_title.plot.bar()
```

```python
# BONUS 2: Bar chart of average salary by title

# Define the query
query_into_pandas = pd.read_sql("SELECT * FROM Salaries INNER JOIN Titles ON Salaries.emp_no = Titles.emp_no", connection)

# Store the query
avg_salaries_by_title = query_into_pandas.groupby(["title"]).mean()["salary"]

# Plot the Barchart
title_list = avg_salaries_by_title.index
avg_salary_for_title = avg_salaries_by_title.values

x_axis = np.arange(len(title_list))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, title_list, rotation=90)
plt.title("Average Salary By Title")
plt.xlabel("Title")
plt.ylabel("Average Salary ($)")
plt.grid(axis="y")
chart = plt.bar(x_axis, avg_salary_for_title, color="b", alpha=0.5, align="center")

for i in chart:
    yvalue = i.get_height()
    plt.text(i.get_x(), yvalue + 500, "${:,.2f}K".format(yvalue/1000))
    
plt.show()
```
